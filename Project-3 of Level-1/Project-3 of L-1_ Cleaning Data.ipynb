# Missing data handling

```python
# Are there any missing values in the dataset?

df.isnull().sum() # tells no of missing values in each column in the pandas series format
```

**These columns has missing values: `name`, `host_name`, `last_review` , `reviews-per-month`**

```python
# Check for missing values in the selected columns
missing_indexes = df[df[['name', 'host_name', 'last_review', 'reviews_per_month']].isnull().any(axis=1)].index

# Print the indexes
print(missing_indexes.tolist())
```

```python
# Get rows where any of the specified columns have missing values
missing_rows = df[df[['name', 'host_name', 'last_review', 'reviews_per_month']].isnull().any(axis=1)]

# Extract the 'id's of those rows
missing_ids = missing_rows['id'].tolist()

# Print the missing IDs
print(missing_ids)
```

```python
# Fill missing 'name' with a placeholder
df['name'].fillna('Unknown', inplace=True)

# Fill missing 'host_name' with a placeholder
df['host_name'].fillna('Unknown', inplace=True)

# Fill missing 'last_review' with the median date or a placeholder date
# Option 1: Fill with median date
# df['last_review'] = pd.to_datetime(df['last_review'])  # Convert to datetime
# median_date = df['last_review'].median()
# df['last_review'].fillna(median_date, inplace=True)

# Option 2: If you want to use a fixed date instead
df['last_review'].fillna('2020-01-01', inplace=True)

# Fill missing 'reviews_per_month' with 0 (assumes no reviews if missing)
df['reviews_per_month'].fillna(0, inplace=True)
```

# Outlier detection

```python
from scipy import stats
import numpy as np
import pandas as pd

# Calculate Z-scores
z_scores = np.abs(stats.zscore(df['reviews_per_month']))

# Get outlier indexes (Z > 3)
outliers_z = df[z_scores > 3]

print("Z-score Outliers:\n", outliers_z[['id', 'reviews_per_month']])
```

```python
# Calculate Z-scores
z_scores = np.abs(stats.zscore(df['availability_365']))

# Get outlier indexes (Z > 3)
outliers_z = df[z_scores > 3]

print("Z-score Outliers:\n", outliers_z[['id', 'availability_365']])
```

```python
# Calculate Z-scores
z_scores = np.abs(stats.zscore(df['calculated_host_listings_count']))

# Get outlier indexes (Z > 3)
outliers_z = df[z_scores > 3]

print("Z-score Outliers:\n", outliers_z[['id', 'calculated_host_listings_count']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: set style
sns.set(style="whitegrid")

# Plot boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['reviews_per_month'])
plt.title('Box Plot of Reviews Per Month')
plt.xlabel('Reviews per Month')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['reviews_per_month'], bins=50, kde=True)
plt.title('Distribution of Reviews per Month')
plt.xlabel('Reviews per Month')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['id'], df['reviews_per_month'], alpha=0.5)
plt.title('Scatter Plot: ID vs Reviews per Month')
plt.xlabel('ID')
plt.ylabel('Reviews per Month')
plt.show()
```

```python
# Recalculate IQR bounds
Q1 = df['reviews_per_month'].quantile(0.25)
Q3 = df['reviews_per_month'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create outlier flag
df['is_outlier'] = (df['reviews_per_month'] < lower_bound) | (df['reviews_per_month'] > upper_bound)

# Plot with outliers in red
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='id', y='reviews_per_month', hue='is_outlier', palette={True: 'red', False: 'blue'}, alpha=0.6)
plt.title('Outlier Detection: Reviews per Month')
plt.xlabel('ID')
plt.ylabel('Reviews per Month')
plt.legend(title='Outlier')
plt.show()
```
